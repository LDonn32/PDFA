## Some Concepts


Mysql can have a number of databases

Each database can have a number of tables. 

Tables have columns (which have a data-type)

- int

- varchar

- bool

- date, time, datetime

Tables also have rows that contain the data that we can perform CRUD on:

- Create (insert)

- Read (select)

- Update

- Delete

## Some Commands

Note: ALL commands end with a semicolon ;

show databases;

create database test_db;

use test_db ( this is the only command where we don't need to add ; but we can to be safe)

create table test_table(id int, 
    name varchar(250)
    );

insert into test_table (id,name) values (1, 'joe');

select * from test_table;

update test_table set name='blah' where id = 1;

delete from test_table where value = 1;
